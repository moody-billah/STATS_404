Importing libaries.

```python
import sqlite3
import pandas as pd
```

Connecting to the 'chinook' database.

```python
conn = sqlite3.connect('chinook.db')
```

Writing the query to get the top 3 most purchased (by revenue) artists in California (CA) each year.

```python
query = """
select *
from (
select Year, Name, TotalRevenue, rank() over(partition by Year order by TotalRevenue desc) as Position
from (
select Year, ArtistId, Name, sum(Revenue) as TotalRevenue
from (
select strftime('%Y',i.InvoiceDate) as Year, al.ArtistId, ar.Name, ii.UnitPrice*ii.Quantity as Revenue 
from invoice_items as ii
join invoices as i
on ii.InvoiceId = i.InvoiceId
join customers as c
on i.CustomerId = c.CustomerId
join tracks as t
on ii.TrackId = t.TrackId
join albums as al
on t.AlbumId = al.AlbumId
join artists as ar
on al.ArtistId = ar.ArtistId
where c.State = 'CA'
) group by Year, ArtistId
)
) where Position <= 3
order by Year, Position
"""
```

Running the query and showing the results.

```python
df = pd.read_sql_query(query, conn)
df
```
